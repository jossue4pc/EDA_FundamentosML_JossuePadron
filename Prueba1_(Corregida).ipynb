# Identificacion de valores atipicos

```python
# Calcula los cuartiles
Q1 = df['MonthlyIncome'].quantile(0.25)
Q3 = df['MonthlyIncome'].quantile(0.75)
IQR = Q3 - Q1

# Define los límites para valores atípicos
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Identifica los valores atípicos
valores_atipicos = df[(df['MonthlyIncome'] < limite_inferior) | (df['MonthlyIncome'] > limite_superior)]

# Imprime los valores atípicos
print(valores_atipicos[['EmployeeNumber', 'JobRole', 'MonthlyIncome']])
```

Se uso el metodo IQR para poder identificar los valores atipicos en la variable de ingreso mensual, estos valores representan empleados con ingresos mensuales que son inusualmente altos en comparacion al resto, se observa que muchos de los valores atipicos corresponden a empleados con puestos de trabajo como Manager o director de investigacion lo que sugiere que estos puestos tienden a tener sueldos mas altos

# Estadistica inferencial:

```python
from scipy import stats

# Supongamos que 'HourlyRate' es la variable de sueldo por hora
media_esperada = 60
sample = df['HourlyRate'].dropna()

# Test t de una muestra
t_stat, p_value = stats.ttest_1samp(sample, media_esperada)

print(f'Estadístico t: {t_stat:.3f}')
print(f'Valor p: {p_value:.3f}')

if p_value < 0.05:
    print("Se rechaza H0: La media es significativamente distinta a 60.")
else:
    print("No se rechaza H0: No hay evidencia suficiente para decir que la media es distinta de 60.")
```

# Algebra lineal

```python
import numpy as np

# Seleccionamos dos variables numéricas
X = df[['MonthlyIncome', 'YearsAtCompany']].dropna().values

# Convertimos a matriz y aplicamos una transformación lineal
X_matrix = np.array(X)

# Producto punto entre columnas
producto = np.dot(X_matrix[:, 0], X_matrix[:, 1])
print(f'Producto punto entre MonthlyIncome y YearsAtCompany: {producto:.2e}')

# Norma (longitud del vector) de la primera columna
norma = np.linalg.norm(X_matrix[:, 0])
print(f'Norma de MonthlyIncome: {norma:.2f}')
```

# Comparacion: Media vs Mediana segun distribucion


En distribuciones simetricas, la media y la mediana tienden a ser iguales, por lo que usar la media es representativo.

En distribuciones asimétricas (sesgadas), la mediana es más robusta porque no se ve afectada por valores extremos .Por ejemplo, si algunos sueldos son extremadamente altos, la media se distorsiona, pero la mediana sigue siendo representativa de la mayoría.

En el analisis, observamos que "HourlyRate" tiene un sesgo negativo, lo que sugiere que la mediana puede ser más confiable que la media para representar el sueldo típico.
